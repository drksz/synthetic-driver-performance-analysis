# Dataset Information 

### Description 
The `driver-deliveries.csv` dataset is a synthetic dataset composed of simulated delivery records for an undisclosed logistics company in Metro Manila, Philippines. Each observation in the dataset is a unique delivery record containing ETA, driver ratings, weather info., and traffic condition.


### Source
The dataset used was generated using a custom-built Python script that creates artificial ddelivery transaction records. The script was used to generate $600$ unique driver profiles with *internal* parameters *(e.g. friendliness, efficiency)* to influence delivery characteristics *(delays, driver ratings, etc.)*. The script also introduces synthetic variance to the dataset by using various probability distributions suited to the values being generated. The script can be accessed [here](../script/generator.py).
