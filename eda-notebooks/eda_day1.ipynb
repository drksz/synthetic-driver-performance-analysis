# Dataset Information 

### Description 
The `driver-deliveries.csv` dataset is a synthetic dataset composed of simulated last-mile, same-day delivery records of an undisclosed logistics company. Each observation in the dataset is a unique delivery record containing ETA, driver ratings, weather info., and traffic condition.


### Source
The dataset used was generated using a custom-built Python script that creates artificial ddelivery transaction records. The script was used to generate $600$ unique driver profiles with *internal* parameters *(e.g. friendliness, efficiency)* to influence delivery characteristics *(delays, driver ratings, etc.)*. The script also introduces synthetic variance to the dataset by using various probability distributions suited to the values being generated. The script can be accessed [here](../script/generator.py).

# Import

### Necessary Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette("muted")
```

### The Dataset

```python
deliveries = pd.read_csv('../data/driver-deliveries.csv')
```

```python
deliveries.head(3)
```

# Data Validation

Before performing any formal analysis, it's a good practice to perform preprocessing steps. 

### Null Value Checks

```python
deliveries.isna().sum()
```

Checking for artificial/placeholder nulls ('N/A','MISSING','UNKNOWN', etc.) were skipped since we already know that the dataset generator does not generate such values. It is an important step to be done when working with other datasets, however. 

### Data Types
Checking data types for each feature. Some features are casted to their appropriate type for proper formatting and to save memory (categoricals). The data dictionary above is used as a guide. 

```python
deliveries.info()
```

During this time, feature cardinality can also be checked. Parts of the data (such as literal feature values) can also be 'standardized' to make them more readable or easier to work with.

Since there are categoricals, we check the unique values for each and standardize them first before casting them to categorical data type.

```python
categoricals = ['priority','vehicle_type','weather','traffic_cond']

for cat in categoricals:
    print("-"*25, f"\nFeatures for column \'{cat}\'")
    print(deliveries[cat].unique())
```

```python
# transform feature values to make them easier to read 
deliveries['priority'] = deliveries['priority'].replace(
    {'EXP':'Express', 'STD':'Standard'}
)
deliveries['weather'] = deliveries['weather'].replace(
    {'CLR':'Clear', 'RAIN':'Rainy'}
)

deliveries['vehicle_type'] = deliveries['vehicle_type'].str.title()
deliveries['traffic_cond'] = deliveries['traffic_cond'].str.title()
```

```python
# then cast to the target data types
deliveries['date'] = pd.to_datetime(deliveries['date'])

for cat in categoricals:
    deliveries[cat] = deliveries[cat].astype('category')

deliveries['traffic_cond'] = deliveries['traffic_cond'].cat.reorder_categories(['Light','Medium','Heavy'], ordered=True)
```

```python
# finally, check the dtypes
deliveries.info()
```

### Visualizing Distributions and Outlier Detection
The numerical features are plotted using different graphical visualizations to identify extreme values. 

Since the rating columns (pkg_care, delivery_spd, etc.) are constrained between 1.0 and 5.0 inclusive, we can be sure no extremes occur there. This leaves us with only four numerical features to check for outliers.

```python
numericals = [
    'delivery_distance',
    'idle',
    'arrival_est',
    'arrival_act'
]

fig, ax = plt.subplots(2, 2, figsize=(12,10))

ax_row = 0
ax_col = 0

for i, num in enumerate(numericals):

    if ax_col > 1:
        ax_col = 0
        ax_row += 1

    sns.boxplot(
        data=deliveries,
        x=num,
        ax=ax[ax_row, ax_col]
    )

    ax_col += 1

ax[0,0].set_xlabel('delivery_distance [km]')
ax[0,1].set_xlabel('idle [mins.]')
ax[1,0].set_xlabel('arrival_est [mins.]')
ax[1,1].set_xlabel('arrival_act [mins.]')

plt.show()
```

Note that these numeric features follow the same right-skewed distribution, where the bulk of the data is on the left and extreme datapoints along the right tail. We then check the statistical attributes of these features. 

```python
for num in numericals:
    print("-"*25)
    print(deliveries[num].describe().loc[['mean','min','25%','50%','75%','max']])
```

Using this information, we can then analyze why such values occurred. 

The max `delivery_distance` in kilometers is about 45. For intercity (or even intracity for large ones) deliveries, this is likely a reasonable value. Although this depends heavily on the specific geographic location, it can be left in the data because it may be the case that this specific delivery's destination is at the farthest point of a large city. 

The same goes for `idle` time. The highest idle time recorded in the dataset was 35 minutes. It does sound an absurdly long time for a single stop, but this feature tracks **total** idle time. This includes idle time from instances such as multiple traffic stops, toll queues, etc. `arrival_est` is also another consideration. Unless the GPS or route planning system the company uses was not functioning properly, we could treat the extremes here as erroneous observations (we cannot be sure of that though). However, we can make a judgment call based on the extremes. An estimated arrival time of just above 2 hours (140 mins.) is plausible, if we assume these deliveries were of intercity distance and/or the general road and weather conditions were not that great. Hence, we leave the values untouched. 

Finally, `arrival_act`'s summary statistics suggest that deliveries took over 4 hours (244 mins.), which means that there exists a certain delivery instance which had a maximum delay of almost two hours. Again, this is largely business context-dependent, since 'reasonable' delays usually depend on  the type of cargo (perishables, high-value items), geographic location, and traffic condition. 

For this, we will check if such late deliveries consistently happen or not by evaluating it's proporition to the rest of the data.

First, we make a judgment call and define what could be the reasonable delivery delay threshold in minutes. Assuming the deliveries are done from local hubs in the city, recipients would expect to get their packages early, generally tolerating delays of up to an hour. We'll set the threshold to 60 and count the records that had delays of over an hour. 

```python
threshold = 60

mask = (deliveries['arrival_act'] - deliveries['arrival_est']) > threshold
extreme_delay = len(deliveries.loc[mask, :])
delay_prop = (extreme_delay/len(deliveries)) * 100

print(f"Ther are {extreme_delay} ({(round(delay_prop, 2))})% deliveries with delays > 1h.")
```

From here, it appears that barely 0.8% of the delivery records were extremely delayed. These delays might have been caused by occurrences such as roadworks/accidents or vehicle breakdowns. Needless to say, these observations may be true variations. Removing such values usually require further justification. If our goal was to use this dataset to train a model, where that model will be predicting on unseen non-extreme data, then it is deal to drop such extreme values. But since our focus is performing EDA and extracting insights from the data, we can leave it in to keep it representative of the population. 

### Distribution Observations

With extreme values out of the way, we can focus on the details of the data distributions. This time, we use histograms since it gives us a visual representation of what the spread of the data looks like and the presence of more than one peak (multimodality), unlike boxplots which only visualize the range of values. 

```python
fig, ax = plt.subplots(2, 2, figsize=(10,10))

ax_row = 0
ax_col = 0

for i, num in enumerate(numericals):

    if ax_col > 1:
        ax_col = 0
        ax_row += 1

    sns.histplot(
        data=deliveries,
        x=num,
        ax=ax[ax_row, ax_col]
    )

    ax_col += 1

plt.show()
```

##### Note the following:
- All histograms here are right-skewed. This is due to high values (discussed above) pulling the distribution tails to the right. 

- The `idle` histogram appears to be bimodal. That is, there is a shorter peak near the zero mark. While this may look unusual, recall that a vehicle needs to be stationary for a certain amount of minutes for the system to consider it as idling. This suggests that while most deliveries have a median idle time of about 9 minutes, some deliveries were seamless, only being delayed by minor factors such as red traffic lights, stop signs, etc. 

- There is a subtle peak at the rightmost part of the `arrival_est` distribution's tail. This likely indicates that there is a relatively high volume of deliveries being made to the farthest reaches of a city/region. 

### Feature Engineering
Looking at the data once more, it appears that some columns can be used to calculate new features and extract additional insights. 

- `delay` is the amount of time in minutes a delivery is late(+) or early(-). Given by `arrival_act` $-$ `arrival_est`.
- `on_time` a binary feature; true if `delay`$ <= 0$ and false otherwise.
- `est_veh_spd` is the delivery vehicle's speed in km/h, using `delivery_distance`, `idle`, and `arrival_act`.
- `weekday` a date variable. Useful for checking delivery volume per day. Extracted from `date`.

```python
deliveries['delay'] = deliveries['arrival_act'] - deliveries['arrival_est']
deliveries['on_time'] = np.where(deliveries['delay'] < 0, 1, 0)

deliveries['est_veh_spd'] = (
    deliveries['delivery_distance'] / (deliveries['arrival_act'] - deliveries['idle']) * 60
)

day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

deliveries['weekday'] = deliveries['date'].dt.day_name().astype('category')
deliveries['weekday'] = deliveries['weekday'].cat.reorder_categories(day_order, ordered=True)

# reordering columns for a neater look
col_reorder = ['delivery_id', 
               'driver_id', 
               'date', 
               'weekday',
               'priority', 
               'vehicle_type', 
               'delivery_distance', 
               'idle', 
               'arrival_est', 
               'arrival_act', 
               'delay', 
               'on_time', 
               'est_veh_spd',
               'attitude', 
               'pkg_care', 
               'responsiveness', 
               'delivery_spd', 
               'weather', 
               'traffic_cond'
    ]

deliveries = deliveries.loc[:, col_reorder]

deliveries.head(3)
```

### Correlations

Now we can start checking which features correlate with one another.

We start by using a heatmap to check the Pearson correlation of each numeric feature with each other.

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.heatmap(
    deliveries.corr(numeric_only=True),
    ax=ax,
    annot=True
)

plt.show()
```

We direct our attention to the non-diagonal tiles. From them, the following can be observed: 


1. `idle` time is moderately correlated with `delay`. The more a delivery vehicle sits in traffic, the higher the chance of it missing delivery schedule. 

2. There is a moderate negative correlation between `est_veh_spd` and `delay`. It means that faster moving delivery vehicles minimize delays. 

3. Notice on the lower right portion of the map that the pairwise correlations for the driver ratings are weak (`attitude`, `pkg_care`, `responsiveness`, `delivery_spd`). For instance, a customer may poorly rate a driver in terms of perceived delivery speed but may give them a good package care rating. Hence, there is no meaningful correlation between them.

4. There is a strong negative correlation between `delay` and `delivery_spd`. Little to no delays often grant delivery drivers higher perceived delivery speed ratings. This also shows how powerful feature engineering is, as we would not have uncovered this correlation with only `arrival_est` and `arrival_act`.


Correlation coefficients usually are not enough to verify true correlation alone. Hence, we use scatterplots to view the structure of the relationship for pairwise correlations listed in 1, 2, and 4. 

```python
fig, ax = plt.subplots(1, 3, figsize=(17, 6))

corr_cols = [
    ['idle','delay'],
    ['est_veh_spd','delay'],
    ['delay', 'delivery_spd']
]

for idx, corr in enumerate(corr_cols):

    # sampling only 1000 observations to avoid crowding the plot while preserving structure

    sns.scatterplot(
        data=deliveries.sample(1000, random_state=42),
        x=corr[0],
        y=corr[1],
        ax=ax[idx]
    )

plt.suptitle('Pairwise correlations')
plt.tight_layout()
plt.show()
```

Using the scatterplots, we immediately see a non-linear trend for the `est_veh_spd` and `delay` pairwise correlation. Instead of a linear trend, it is more like an exponential decay. This, however, does not mean that the correlation is weak. However, estimators such as linear regression models would give inaccurate predictions with such data. 

The other two plots show strong positive and negative correlations, confirming the relationships mentioned above. 

### Aggregates

Now that the correlations have been taken care of, we can examine counts and other general insights. 

```python
fig, ax = plt.subplots(2, 2, figsize=(17,11))

data=deliveries
on_time_map = deliveries['on_time'].map({1:'Yes',0:'No'})


sns.countplot(
    data=data,
    x= on_time_map,
    order=['Yes','No'],
    hue='priority',
    ax=ax[0,0],
)

sns.histplot(
    data=data,
    x='weekday',
    hue=on_time_map,
    ax=ax[0,1],
    multiple='stack',
    shrink=0.8,
)

sns.countplot(
    data=data,
    y='vehicle_type',
    hue=on_time_map,
    ax=ax[1,0],
)

sns.histplot(
    data=data,
    y='traffic_cond',
    hue=on_time_map,
    ax=ax[1,1],
    multiple='stack',
    shrink=0.8,
)

ax[0,1].set_ylim(0, 19000)
ax[1,0].set_xscale('log')
ax[1,0].set_xlabel('count (logscale)')
ax[1,1].set_xlabel('count (logscale)')
ax[1,1].set_xscale('log')

plt.tight_layout()
plt.show()
```

The 4-plot figure shows aggregates of deliveries. Note that some axes used logarithmic scales to enhance plot readability. 


The first clustered bar chart shows the total deliveries according to shipping priority. It appears that the number of delayed deliveries dominate the dataset. Moreover, regardless of delay, *standard* shipping is the more popular shipping priority of choice made by customers. 

To the right of that plot, is a stacked bar chart showing the volume of deliveries made for each day of the week. Delivery volume appears to be roughly the same throughout the week, with weekends usually having the highest number of on-time deliveries.

The clustered bar chart on the lower left portion of the figure shows deliveries made per vehicle type. The plot suggests that the company's fleet is mostly composed of motorcycles than four-wheel vehicles. An advantage held by motorcycles is that they can weave and maneuver through traffic much more easily, which is a likely explanation as to why this group has the highest number of on-time deliveries. 

Finally, the last stacked bar chart shows the volume of deliveries for each traffic condition. On-time delivery counts seem to tie at times of light and medium traffic congestion but are extremely low during heavy traffic, which is expected. 

```python
fig, ax = plt.subplots(1, 2, figsize=(18,6))

sns.boxplot(
    data=deliveries,
    y='weather',
    x='est_veh_spd',
    ax=ax[0]
)
sns.boxplot(
    data=deliveries,
    y='traffic_cond',
    x='idle',
    ax=ax[1]
)

plt.show()
```

These boxplots compare estimated vehicle speed according to weather, and idle time based on traffic condition. 

The first boxplot compares weather and vehicle speed. Notice how the whiskers and the center are pulled slightly more towards the left, indicating that drivers tend to be more careful with their speed during wet road conditions. 

The other boxplot describes the central tendencies of idle time according to traffic. As expected, delivery vehicles caught in heavy traffic idle for longer periods of time compared to vehicles in periods of light traffic. 

```python
fig, ax = plt.subplots(2, 2, figsize=(18,13))

sns.boxplot(
    data=deliveries,
    x='attitude',
    ax=ax[0,0]
)
sns.boxplot(
    data=deliveries,
    x='pkg_care',
    ax=ax[0,1]
)

sns.boxplot(
    data=deliveries,
    x='delivery_spd',
    ax=ax[1,0]
)

sns.boxplot(
    data=deliveries,
    x='responsiveness',
    ax=ax[1,1]
)

plt.show()
```

This collection of boxplots illustrates the distribution of driver ratings across deliveries. 

Among the rating features, delivery speed shows the narrowest interquartile range, indicating that drivers consistently meet customer expectations regarding delivery timeliness. In contrast, responsiveness displays the greatest variability among the four, suggesting that the quality of driver–customer communication differs more noticeably across deliveries.
